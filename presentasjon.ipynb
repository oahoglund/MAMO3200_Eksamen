## Schrödinger ligningen i 2 + 1 dimensjoner

$$
i\,\frac{\partial u(\mathbf{x}, t)}{\partial t}
= -\frac{\nabla^{2}}{2}\,u(\mathbf{x}, t) + v(\mathbf{x})\,u(\mathbf{x}, t)
= \hat{H}\,u(\mathbf{x}, t), \qquad
\hat{H} = -\frac{\nabla^{2}}{2} + v(\mathbf{x})
$$
$$p(\mathbf{x}; t) = |u(\mathbf{x}, t)|^{2} = u^{*}(\mathbf{x}, t)\,u(\mathbf{x}, t)$$

$$\int_{-\infty}^{\infty} p(x;t)\, dx
= \int_{-\infty}^{\infty} |u(\mathbf{x}, t)|^{2}\, dx
= 1$$

- Har satt planck reduserte konstant og masse til 1
- Skriver også bølgefunksjonen her som u
- $\nabla^2$ Laplacian
- $\hat{H}$ Hamiltonian
- $v(\mathbf{x})$ Potensialfunksjonen
- $p(\mathbf{x}; t)$ sannsynelighetstettheten

$$\nabla	^2 = \frac{\partial ^2u}{\partial x^2} + \frac{\partial ^2u}{\partial y^2}$$

$$ \nabla^2 =

\end{pmatrix}
$$

$$
A =\frac{1}{h^2}
\begin{pmatrix}

## FTCS, euler forward
Bruker forward time:
$$
\frac{\partial u}{\partial t} = \frac{u^{n+1}-u^n}{\Delta t}
$$

Setter sammen med hamiltonian
$$
i\frac{u^{n+1}-u^n}{\Delta t} = Hu^n
$$
så
$$
u^{n+1} = u^n - i\Delta t Hu^n 
$$

## Leapfrog
Har dette skjema:
$$u^{n+1}=u^{n-1}-2i\Delta tHu^n$$

For å finne $u^1$ så bruker vi rk2 med skjema:
$$
\begin{aligned}
k_1 &= -i\, (H u^n), \\
k_2 &= -i\, H\left(u^n + \tfrac{1}{2}\, \Delta t\, k_1\right), \\
u^{n+1} &= u^n + \Delta t\, k_2 .
\end{aligned}
$$


### Stabilitet
- Leapfrog er ment til å være stabil for:
$$\Delta t \leq \frac{1}{v_0+1/h^2}\approx 10^{-10}$$
- (Vi hadde $\Delta t = 8 \times 10^{-7}$)
- Svarer til $N= \frac{T_{end}}{\Delta t}+1 \approx 10^{10}\times 8\times 10^{-3}=8\times 10^7$ iterasjoner

- For lavere v0, $$v_0=1\times 10^4$$
- $$\Delta t \leq \frac{1}{v_0+1/h^2}\approx 5\times 10^{-5}$$
- Svarer til $N\approx 160$

- Eksperimentelt så er grensen med $N = 368$, under så var den ustabil

## Runge-Kutta 4
Har følgende skjema
$$
\begin{aligned}
k_1 &= -i\, H u^n, \\
k_2 &= -i\, H\left(u^n + \tfrac{1}{2}\Delta t\, k_1\right), \\
k_3 &= -i\, H\left(u^n + \tfrac{1}{2}\Delta t\, k_2\right), \\
k_4 &= -i\, H\left(u^n + \Delta t\, k_3\right), \\
u^{n+1} &= 
u^n + \frac{\Delta t}{6}\left(k_1 + 2k_2 + 2k_3 + k_4\right).
\end{aligned}
$$

### Stabilitet
- Leapfrog er ment til å være stabil for:
$$\Delta t \leq \frac{2\sqrt{2}}{v_0+1/h^2}\approx 10^{-10}$$
- Svarer til $N=8\times 10^7$ iterasjoner

- For lavere v0, $$v_0=1\times 10^4$$
- $$\Delta t \leq \frac{2\sqrt{2}}{v_0+1/h^2}\approx 1,4\times 10^{-4}$$
- Svarer til $N\approx 58$

- Stabilitetskravet kommer fra chatgpt

- Eksperimentelt så er grensen med $N = 128^*$, under så var den ustabil

## Crank-Nicholson
Finite differance i tid:
$$
\frac{\partial u}{\partial t} = \frac{u^{n+1}-u^n}{\Delta t}
$$

Høyre siden så bruker vi gjennomsnitt av eksplisitt og implisitt skjema:
$$
\frac{Hu^{n+1}+Hu^n}{2}
$$
Vi setter inn alt i ligningen og får:
$$
(I+\frac{i\Delta t}{2}H)u^{n+1}=(I-\frac{i\Delta t}{2}H)u^n
$$

### For å løse dette:
